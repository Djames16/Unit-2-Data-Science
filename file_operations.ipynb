Using your knowledge of Python and Matplotlib write your code to transform and graph the data below.

After your code create a new Markdown block to describe your analysis

````python
import matplotlib.pyplot as plt
Text= open('states.txt')
States=["AL", "AK","AZ","AR", "CA"]
st_AL=0
for x in Text.readlines():
    if x == "AL\n":
        st_AL+=1


height=[st_AL]

plt.bar(States,height)
plt.xlabel("States")
plt.ylabel("Number")
plt.title("How many states?")
````
