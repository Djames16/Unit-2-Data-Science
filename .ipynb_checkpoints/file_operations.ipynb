Using your knowledge of Python and Matplotlib write your code to transform and graph the data below.

After your code create a new Markdown block to describe your analysis

````python
import matplotlib.pyplot as plt
Text= open('states.txt')
States=["AL", "AK","AZ","AR", "CA"]
st_AL=0
st_AK=0
st_AZ=0
st_AR=0
st_CA=0

for x in Text.readlines():
    if x == "AL\n":
        st_AL+=1
    elif x == "AK\n":
        st_AK+=1
    elif x == "AZ\n":
        st_AZ+=1
    elif x == "AR\n":
        st_AR+=1
    elif x == "CA\n":
        st_CA+=1


height=[st_AL,st_AK,st_AR,st_AZ,st_CA]

plt.bar(States,height)
plt.xlabel("States")
plt.ylabel("Number")
plt.title("How many states?")
````

I can see that the y axis has the number of states in the increments based on the highest number of states. I also see the x axis holds the 5 states that I have listed and shows the bar based on the how many states that is listed.
